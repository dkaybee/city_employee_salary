```python
# We are going to be working with a PDF file to extract city if Saskatoon employee salary and positions information of

# We need to install the necessary modules required
!pip install -q tabula-py
import tabula
import pandas as pd

```

```python
path= "/content/drive/MyDrive/city_of_saskatoon_public_accounts_2020.pdf"

#use the .read_pdf to look at the 
#tables = tabula.read_pdf(path, pages="7", lattice=False, pandas_options={'header':None})
```

```python
tabula.convert_into(path,"city_sask_salary.csv", pages="7-69", stream=True)
```

```python
# add column headers manually as they were omitted from tabula

salary_df=pd.read_csv("city_sask_salary.csv", names=['Employee_name', 'Position_title', 'nil_1', 'Total_salary', 'nil_2','Regular_salary', 'nil_3','Retro_pay', 'nil_4','Car_allowance','nil_5','Overtime'])
salary_df.head(5)
```

```python
#drop the columns not needed

salary_df=salary_df.drop(columns=['Employee_name','nil_1', 'nil_2','nil_3','nil_4','nil_5'])
salary_df.head(5)
```

```python
#explore your data, check the datatype

salary_df.info()
```

```python
# We remove ',' so we can analyse the data

salary_df.replace(',','', regex=True, inplace=True)

# replace and remove space " "
salary_df.replace(' ','', regex=True, inplace=True)
salary_df.info()
```

```python
#Convert the monetary columns to numeric datatype
# we use the .apply method

salary_df['Total_salary']=salary_df['Total_salary'].apply(pd.to_numeric, errors='ignore')
salary_df['Regular_salary']=salary_df['Regular_salary'].apply(pd.to_numeric,errors='ignore')
salary_df['Retro_pay']=salary_df['Retro_pay'].apply(pd.to_numeric,errors='coerce')
salary_df['Car_allowance']=salary_df['Car_allowance'].apply(pd.to_numeric,errors='coerce')
salary_df['Overtime']=salary_df['Overtime'].apply(pd.to_numeric,errors='coerce')
```

```python
# Confirm the datatype has changed to int

salary_df.info()
```
