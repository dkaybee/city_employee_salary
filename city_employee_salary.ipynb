```python
# We are going to be working with a PDF file to extract city if Saskatoon employee salary and positions information of

# We need to install the necessary modules required
!pip install -q tabula-py
import tabula
import pandas as pd

```

```python
path= "/content/drive/MyDrive/city_of_saskatoon_public_accounts_2020.pdf"

#use the .read_pdf to look at the 
tables = tabula.read_pdf(path, pages="7", lattice=False, pandas_options={'header':None})
print(tables)
```

```python
tabula.convert_into(path,"city_sask_salary.csv", pages="7-69", stream=True)
```

```python
salary_df=pd.read_csv("city_sask_salary.csv", names=['Employee_name', 'Position_title', 'nil_1', 'Total_salary', 'nil_2','regular_salary', 'nil_3','Retro_pay', 'nil_4','Car_allowance','nil_5','Overtime'])
salary_df.head(5)
```

```python
salary_df=salary_df.drop(columns=['nil_1', 'nil_2','nil_3','nil_4','nil_5'])
```
